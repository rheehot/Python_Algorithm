```python
# 노드는 데이터와 , 가르키는 참조 값으로 구성
class Node:
    
    def __init__(self,data):
        self.data = data
        self.Next = None
        
class LinkedList:
    
    def __init__(self):
        start_point = Node("start")
        self.head = start_point
        self.tail = start_point
        
        self.current = None
        self.before = None
        
        self.data_count = 0
        
    # head 는 맨 앞
    # before 현재위치 전
    # current 현재위치
    # tail 맨 뒤
    # data_count 데이터 갯수
    
    # 위의 생성자를 구현하고, 추가, 삭제, 맨앞 노드 검색, 다음 노드 검색 을 할 수 있는 함수들을 만들어야 함
    # 추가 - append , 삭제 - delete 맨 앞 노드 검색 - first , 다음 노드 검색 - next
    def append(self,data):
        new_node = Node(data)
        self.tail.next = new_node # 사실 이게 왜 있는지 모르겠다.
        self.tail = new_node
        
        self.data_count += 1
    
    def delete(self):
        # pop_data = self.current.data
        
        if self.current is self.tail:
            self.tail = self.before
        
        self.before.next = self.current.next
        self.current = self.before
        self.data_count -= 1
        
        # return pop_data 삭제를 확인하기 위해서 만든 듯
        
    def first(self):
        if self.data_count == 0:
            return None
        # 제일 처음의 뒤의 데이터, 시작한 값을 불러와서 그 뒷값을 보여줘서 리스트의 시작부분을 보여주네
        self.before = self.head
        self.current = self.head.next
        
        return self.current.data
    
    def next(self):
        if self.current.next == None:
            return None
        
        # 이거는 다음 값으로 넘어가기 위한 next
        self.before = self.current
        self.current = self.current.next # 여기 이 next는 솔찍히 뭐가 뭔지 모르겠다.
        # next가 다음 값을 가르키고 있다면 모르겠지만 Append에서도 next 값을 뭘 가르치고 있는지 이해가 안감
        
    def size(self):
        return self.data_count
    

if __name__ =="__main__":
    test_list = LinkedList()
    test_list.append(50)
    test_list.append(10)
    test_list.append(30)
    test_list.append(20)
    test_list.append(40)
    
    print("first :",test_list.first())
    print("next :",test_list.next())
    # 일단.. 잠깐 여기까지만 올려놔 보자..
```
