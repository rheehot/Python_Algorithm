```python
if __name__=="__main__":
    file_n = int(input())
    file_size = list(map(int,input().split()))
    cluster_size = int(input())
    count = 0
    for i in range(file_n):
        if file_size[i] % cluster_size == 0:
            file_size[i] //= cluster_size
        else:
            file_size[i] = file_size[i] // cluster_size + 1
        count += file_size[i]
    print(cluster_size * count)
```

```python
if __name__=="__main__":
    call_n = int(input())
    call_list = list(map(int,input().split()))
    Y, M = 0, 0
    for i in range(call_n):
        Y += (call_list[i] // 30 + 1) * 10
        M += (call_list[i] // 60 + 1) * 15
    if Y < M:
        print("Y",Y)
    elif Y == M:
        print("Y M",Y)
    else:
        print("M",M)
```

```python
def a(s_list):
    s_list.sort()
    s_list = s_list[::-1]
    return s_list

if __name__=="__main__":
    s_list = list(map(int,input().split()))
    s_list = a(s_list)
    print(s_list)
```

```python
def factorial(n):
    factorial = 1
    for i in range(1,n+1):
        factorial *= i
    return factorial

if __name__=="__main__":
    n,m = map(int,input().split())
    result = factorial(m)
    result /= (factorial(n)*factorial(m-n))
    print(int(result))
```
